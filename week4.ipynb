**Task: Create a function that prints rectangle information with length and width.**

`An example output:`

**Output:**
```
📐 Rectangle Details 📐
Length: 5 units
Width: 3 units
Area: 15 square units
Perimeter: 16 units
------------------------------

````python
# Solution:
def rect(length, breadth):
    area = length * breadth
    perimeter = 2 * (length + breadth)
    print("Rectangle Details ")
    print(f"Length: {length} ")
    print(f"Width: {breadth} ")
    print(f"Area: {area} ")
    print(f"Perimeter: {perimeter} ")
    print("=" * 30)
rect(5, 3)
````

Create a function that calculates the area of a circle and returns it.

`(Area = 3.14 × radius × radius)`

**Output:**
```
Circle area: 78.5
Larger circle area: 314.0
Combined area: 78.26

````python
# Solution:
import math
def area(radius):
    return math.pi * radius * radius
area1 = area(4)
area2 = area(9)
result = area1 + area2
print(f"Circle area: {area1}")
print(f"Larger circle area: {area2}")
print(f"Combined area: {result}")
````

````python
import math
def area(radius):
    return math.pi * radius * radius
r1 = int(input("Enter the 1st radius: "))
r2 = int(input("Enter the 2nd radius: "))
area1 = area(r1)
area2 = area(r2)
combined = area1 + area2
print(f"Circle area: {round(area1, 2)}")
print(f"Larger circle area: {round(area2, 2)}")
print(f"Combined area: {round(combined, 2)}")
````

**Task: Create a function that calculates total and average from three game scores and returns both.**

```

Output:

Total: 270, , Average: 90.0
Your total is 280 and average is 93.33333333333333
```

````python
# Solution:
def scores(score1, score2, score3):
    total = score1 + score2 + score3
    average = total / 3
    return total, average
tot1, avg1 = scores(40, 30, 55)
print(f"Total: {tot1}, Average: {avg1}")
tot2, avg2 = scores(100, 90, 90)
print(f"Total is {tot2} and average is {avg2}")
````
