**Palindrome Checker**

**Task: Write a function is_palindrome(text) that returns True if the string is the same when reversed, otherwise False.**

````python
# Solution:
def is_palindrome(text):
    return text == text[::-1]
print(is_palindrome("madam"))   # True
print(is_palindrome("hello"))   # False
````

**Multiple Return Values**

**Task: Create a function circle_measurements(radius) that returns both area and circumference of a circle.**

**Use formulas:**

**Area = π × r²**

**Circumference = 2 × π × r**

````python
# Solution:
import math
def circle(radius):
    area = math.pi * radius ** 2
    circum = 2 * math.pi * radius
    return area, circum
area, circum = circle(7)

print(f"Area: {area}")
print(f"Circum: {circum}")
````

**Task: Create a function that calculates total and average from three game scores and returns both.**

```

Output:

Total: 270, , Average: 90.0
Your total is 280 and average is 93.33333333333333
```

````python
# Solution:
def scores(score1, score2, score3):
    total = score1 + score2 + score3
    average = total / 3
    return total, average
tot1, avg1 = scores(40, 30, 55)
print(f"Total: {tot1}, Average: {avg1}")
tot2, avg2 = scores(100, 90, 90)
print(f"Total is {tot2} and average is {avg2}")
````

**Task: Create a function that finds both the smallest and largest of three numbers.**
```
Output:

Smallest: 5, Largest: 12
Range: 25 to 100
```

````python
# Solution:
def find_min_max(num1, num2, num3):
    smallest = min(num1, num2, num3)
    largest = max(num1, num2, num3)
    return smallest, largest

small, large = find_min_max(5, 12, 8)
print(f"Smallest: {small}, Largest: {large}")

small2, large2 = find_min_max(25, 100, 60)
print(f"Range: {small2} to {large2}")
````

**Task: Create a function that takes a full name and returns first name and last name separately.**

**Output:**
```
Name Amrita Prasad
First name: Amrita, Last name: Prasad
Hello Amrita! Your last name is Prasad.

````python
# Solution:
def split_name(full_name):
    parts = full_name.split()
    first_name = parts[0]
    last_name = parts[-1]
    return first_name, last_name
name = "Anu lakshmi"
first, last = split_name(name)
print(f"Name: {name}")
print(f"First name: {first}, Last name: {last}")
print(f"Hello {first}! Your last name is {last}.")
````

**Task: Create a function that can add any amount of numbers together**

**Task: Create a function that finds the highest number from any amount of scores**

````python
# Solution:

def highest(*scores):
    return max(scores)
print(highest(45, 78, 90, 66))
print(highest(100, 85, 92))
````

**Task: Create a function that calculates the average of any amount of numbers**

````python
# Solution:
def average_numbers(*numbers):
    return sum(numbers) / len(numbers)
print(average_numbers(10, 20))
print(average_numbers(5, 10, 15, 20))
print(average_numbers(1, 2, 3, 4, 5))
````
